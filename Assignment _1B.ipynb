# Predicting the strength of Cement using Regression model in Keras

### Importing required libraries

```python
import pandas as pd
import numpy as np
import keras
from keras.models import Sequential
from keras.layers import Dense
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
```

### Loading the dataset to pandas data frame

```python
concrete_data = pd.read_csv('https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DL0101EN/labs/data/concrete_data.csv')
concrete_data.head()
```

### Checking for null values

```python
concrete_data.isnull().sum()
```

### Splitting the data into target and predictors

```python
Y = concrete_data['Strength']
X = concrete_data.drop('Strength', axis = 1)
n_cols = X.shape[1]
```

### Defining a function of Keras Sequential model with 1 hidden layer

```python
def regression_model():
    model = Sequential()
    model.add(Dense(10, activation = 'relu', input_shape = (n_cols,)))
    model.add(Dense(1))
    
    model.compile(optimizer = 'adam', loss = 'mean_squared_error')
    return model
```

### Defining function to fit the model, predict and return the mean squared error

```python
def mean_se():
    X_train, X_valid, Y_train, Y_valid = train_test_split(X, Y, test_size = 0.3)
    model.fit(X_train, Y_train, validation_data = (X_valid, Y_valid), epochs = 50)
    Y_pred = model.predict(X_valid)
    mse = mean_squared_error(Y_valid, Y_pred)
    print(f'mse = {mse}')
    return mse
```

### Initializing the model

```python
model = regression_model()
```

### Calling the function 50 times and storing the mean squared error in a list

```python
MSE_A = []
for i in range(51):
    MSE_A.append(mean_se())
```

### Calculating mean and standard deviation of all the mean squared errors

```python
mean_mse = np.mean(MSE_A)
sd_mse = np.std(MSE_A)
```

```python
print(f'Mean of mean squared errors is {mean_mse} \n Standard Deviation of mean_sqaured errors is {sd_mse}')
```

### Defining function to split, normalize, fit the model, predict and return the mean squared error

```python
def mean_sen():
    X_train, X_valid, Y_train, Y_valid = train_test_split(X, Y, test_size = 0.3)
    X_train = (X_train - np.mean(X_train)) /  np.std(X_train)
    X_valid = (X_valid - np.mean(X_valid)) / np.std(X_valid)
    model.fit(X_train, Y_train, validation_data = (X_valid, Y_valid), epochs = 50, verbose = 2)
    Y_pred = model.predict(X_valid)
    mse = mean_squared_error(Y_valid, Y_pred)
    print(f'mse = {mse}')
    return mse
```

### Calling the function 50 times and storing the mean squared error in a list

```python
MSE_B = []
for i in range(51):
    MSE_B.append(mean_sen())
```

### Calculating mean and standard deviation of all the mean squared errors

```python
mean_mse = np.mean(MSE_B)
sd_mse = np.std(MSE_B)
```

```python
print(f'Mean of mean squared errors is {mean_mse} \n Standard Deviation of mean_sqaured errors is {sd_mse}')
```
